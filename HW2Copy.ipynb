```python
x, y, z, a, b = np.genfromtxt(r'/Users/michaelmalone/molecules/dsmcer-hw2-michaelm45/ALA2fes.dat', unpack=True)

N = 3000 #number of points for plotting/interpolation


xi = np.linspace(x.min(), x.max(), N)
yi = np.linspace(y.min(), y.max(), N)
zi = scipy.interpolate.griddata((x, y), z, (xi[None,:], yi[:,None]), method='cubic')

fig = plt.figure()
CS = plt.contour(xi, yi, zi, [0, 10, 20, 30, 40, 50, 60, 70])
plt.clabel(CS, inline=1, fontsize=10)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
